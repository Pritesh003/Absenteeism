### We can create 4 groups here:
#### Group I: 1 to 14 (Diseases)
#### Group II: 15 to 17 (Pregnancy/Birth related)
#### Group III: 18 to 21 (Poisoning or other reasons)
#### Group IV: 22 to 28 (Light reasons)

```python
reason_columns.columns.values
```

```python
# we will use .loc method. It is a label-based methed (we have to give column name, unlike .iloc (takes integers)).
# for a given row, if all columns 1-14 are 0, then 'Reason1' = 0,
# else (somewhere in the 14 columns we have a 1) 'Reason1' = 1.
reason_type_1 = reason_columns.loc[:,1:14].max(axis=1)    # we can also use .loc[:,'1':'14']
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:,22:].max(axis=1)
```

```python
reason_type_1
```

### Concatenate the reason types to the dataframe df.

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df
```

### Education, Children and Pets  Column.
#### These columns have categorical data, containing integers. Children and Pets column contains the number of children and pets a person have. They have a numerical meaning. Therefore we will leave these 2 columns untouched and <u>focus on Education column</u>.

```python
df_reason_date_modified['Education'].unique()
```

In Education column of this dataset:<br>
1 -> high school<br>
2 -> graduate<br>
3 -> post graduate<br>
4 -> master or doctorate

```python
df_reason_date_modified['Education'].value_counts()
```

#### There are almost 600 students with high school education, and other higher education combinedly counts to just over 100. So in this case it is more convenient to <u>group all the three categories apart from high school as one category</u>.
#### We will substitute 'high school' with '0' and rest of all with '1'.

```python
df_reason_date_modified['Education'] = df_reason_date_modified['Education'].map({1:0, 2:1, 3:1, 4:1})
# while using map function we should take care of assigning new values to all the existing values.
# or the function will misbehave, causing problems in further analysis.
```
